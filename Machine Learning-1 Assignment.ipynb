Q1: Explain the following with example:
1. Artificial Intelligence
2. Machine Learning
3. Deep Learning

Ans. 
1. Artificial Intelligence: Artificial Intelligence is the broad field of creating machines capable of performing tasks that typically require human intelligence. This includes reasoning, learning, problem-solving, perception, language understanding, and more.

Example: Self driving cars, robots, alexa

2. Machine Learning: Machine Learning is a subset of AI that involves training algorithms to learn from data and make predictions or decisions without being explicitly programmed. It focuses on developing systems that can improve their performance on a task over time with more data.

Example:  Email spam filters. These filters learn to identify spam emails by being trained on a large dataset of emails labeled as "spam" or "not spam." Over time, the filter becomes better at recognizing spam by learning from the patterns in the training data. 

3. Deep Learning: Deep Learning is a specialized subset of ML that uses neural networks with many layers (hence "deep") to analyze various types of data. These neural networks are designed to mimic the human brain's structure and function, enabling the system to learn complex patterns and representations.

Example: Object detection, Image recognition, Chatbots, recommendation system

Q2- What is supervised learning? List some examples of supervised learning

Ans.  Supervised learning is a type of machine learning where an algorithm is trained on labeled data. This means the training dataset includes both the input data and the corresponding correct output. The goal is for the algorithm to learn a mapping from inputs to outputs so that it can predict the output for new, unseen inputs.

In supervised learning, the model makes predictions based on input data and is corrected by comparing its predictions to the actual outputs (labels). This correction process is called "training" and continues until the model achieves a desired level of accuracy.

Example: Predicting house prices based on features like square footage, number of bedrooms, and location.

Q3- What is unsupervised learning? List some examples of unsupervised learning

Ans. Unsupervised learning is a type of machine learning where the algorithm is trained on unlabeled data. Unlike supervised learning, there are no explicit outputs or labels provided. The goal of unsupervised learning is to find hidden patterns or intrinsic structures within the input data.

In unsupervised learning, the model tries to learn the underlying patterns and relationships in the data without any guidance. This is useful for exploratory data analysis, dimensionality reduction, and clustering.

Example: Customer segmentation in marketing, where customers are grouped based on purchasing behavior and demographics.

Q5- What are the main difference between supervised, unsupervised and semi-supervised learning?

Ans: Definition: 

Supervised: Learning from labeled data

Unsupervised: Learning from unlabeled data

Semi-supervised: Learning from a small amount of labeled data and a large amount of unlabeled data

Goal: 

Supervised: Predict outcomes based on input-output pairs

Unsupervised: Find hidden patterns or intrinsic structures in data

Semi-supervised: Improve learning accuracy with limited labeled data

Input data:

Supervised: Labeled data (input-output pairs)

Unsupervised: Unlabeled data

Semi-supervised: Combination of labeled and unlabeled data

Output data:

Supervised: Predictions (classification or regression)

Unsupervised: Groupings or data representations (clusters, associations)

Semi-supervised: Enhanced model performance with limited labeled data

Example:

Supervised: Email spam classification, house price prediction

Unsupervised: Customer segmentation, market basket analysis

Semi-supervised: Image classification with few labeled images and many unlabeled images

Q6- What is train, test and validation split? Explain the importance of each term

Ans. Training Set: The training set is a portion of the dataset used to train the machine learning model. It is the data the model learns from, meaning it adjusts its parameters based on this data to minimize error.

Importance: Learning: The model uses the training set to understand the patterns, relationships, and structure in the data.

Parameter Tuning: The model’s parameters (weights, biases) are optimized based on the training data.

validation Set: The validation set is a portion of the dataset used to tune hyperparameters and evaluate the model during the training phase. It provides an unbiased evaluation of a model fit on the training dataset while tuning the model's parameters.

Importance: Hyperparameter Tuning: Helps in selecting the best model architecture and hyperparameters (e.g., learning rate, number of layers in a neural network).

Model Selection: Allows comparison between different models or algorithms to choose the best performing one.

Overfitting Check: Helps in monitoring and mitigating overfitting by providing feedback on model performance on unseen data during training.

Test Set: The test set is a portion of the dataset used to evaluate the final model after training. It is only used once the model has been trained and validated.

Importance: Performance Evaluation: Provides an unbiased evaluation of the final model’s performance on unseen data.

Generalization Check: Assesses how well the model generalizes to new, unseen data.

Comparison: Used to compare different models or approaches in a fair manner.



Q7- How can unsupervised learning is  used  in anomaly detection?

Ans. Unsupervised learning can be effectively used in anomaly detection to identify unusual patterns or outliers in data that do not conform to expected behavior. Since unsupervised learning does not rely on labeled data, it is particularly useful in scenarios where anomalies are rare and not labeled.

Cluster the data into K clusters. Anomalies are data points that do not fit well into any cluster or belong to small, sparse clusters.

 In network traffic analysis, most data points (normal behavior) will form dense clusters, while anomalous activities (e.g., attacks) will be isolated or form small clusters.
